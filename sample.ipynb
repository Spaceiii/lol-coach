```python
import pandas as pd

dataset = pd.read_csv('data/matchData.csv')

df = pd.DataFrame(dataset, columns=dataset.columns)
```

```python
print(df.head())

print(df.describe())
```

```python
print(df.columns)
```

```python
participants = []
for i in range(10):
    col = [col for col in df.columns if 'participant'+str(i) in col.lower()]
    participants.append(df[col])
    participants[i].columns = participants[i].columns.str.replace('participant'+str(i), '')

participants[0]
```

```python
print(len(participants[0]))

df = df[[col for col in df.columns if 'participant' not in col.lower()]]
df.describe()
```

```python
team_0 = df[[col for col in df.columns if 'team0' in col.lower()]]
team_1 = df[[col for col in df.columns if 'team1' in col.lower()]]

df = df[[col for col in df.columns if col.lower() not in ['team0', 'team1']]]

team_0.columns = team_0.columns.str.replace('team0', '')
team_1.columns = team_1.columns.str.replace('team1', '')

print(team_0, team_1)
print(df)
```
